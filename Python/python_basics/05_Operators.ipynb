# Operators


## Operators helps in applying operation on the given opearands

## Types of operators 

1. Arithmetic operators 


   Addion(+)  \
   Substraction(-) \
   Product or Multiplication(*) \
   Division(/) \
   Modulus(%)\
   Floor(//)\
   Exponent(a**b)

 2. Comparison Operator

    Greater than(>) \
    Less than(<)\
    Greater than or equal to (>=)\
    less than or equal to(<=)\
    Equality check (==)  \
    Not equal to (!=)
    
 3. Logical Operators 

    and---> whether if both conditions are true or not\
    or ---> whether if any of the operation is true or not\
    not -->negation or to inverse the result of the given operation 
    

        

# Arithmetic Operator

```python
a=10
b=3
print(a+b) # Addition
print(a-b)# Subtraction
print(a*b)# Multiplication
print(a/b)# Division
print(a//b)# Floor Division
print(a%b)# Modulo
print(a**b)# Exponentiation
```
