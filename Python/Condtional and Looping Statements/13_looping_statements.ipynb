## Range()

## range() fuctgion can be used  for working as our iterable

## range( ) takes input parameter and the first parameter that the range parameter will be taking as the input will be the initial value you want to print or operate on and then  you will be passing the next input after comma as the parameter as the final value and it can take third parameter as well that will be the step size and that is the differernce between the iterating values 

```python
# print numbers from 1 to 50
# for i in range(initial value,end value+1,step value):
for i in range(1, 51,1):
   print(i)
```

## inthe range there is never neccessity for mentioning the step size and similar if you can want to print the numbers or  you to operate on numbers again there is no necessity passing the initial value as well if and only if your initial value is zero

```python
#numbers from 0 to 100
for i in range(101):
    print(i)
```

```python
# print numbers from 50 to 1 with step value -1
for i in range(50,0,-1):
    print(i)
```

```python
# print numbers from 50 to 0 with step value -5
for i in range(50,-1,-5):
    print(i)
```
