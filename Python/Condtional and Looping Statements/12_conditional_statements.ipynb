# Conditional Statements

## Condtional statements  works like real life scenerio  where if a certain condition is meeting then a statement  of code or block of code gets executed

```python
# Eligibility for drving a car  is meeting certain condition :

Age=int(input("Enter your Age :"))

if Age>=18:
    print("You are eligible for driving a car")

else:
    print("Sorry ! You are not eligible for driving a car")    
```

# Nested Conditional statement

```python
marks = int(input("Enter marks: "))

if marks >= 40:
    if marks >= 75:
        print("Distinction")
    elif marks >= 60:
        print("First Class")
    else:
        print("Pass")
else:
    print("Fail")
```

```python
year = int(input("Enter a year: "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("Leap Year")
        else:
            print("Not a Leap Year")
    else:
        print("Leap Year")
else:
    print("Not a Leap Year")
```
