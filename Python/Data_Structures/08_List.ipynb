```python
## Slicing with step means that you can specify the step size for slicing the list. The syntax for slicing with step is list[start:stop:step]. The step parameter determines how many elements to skip between each element in the slice. For example, if you want to get every second element from a list, you can use a step of 2.
print(list_item[::2])
```

```python
## If you want to get from the end of the list with a step of 3, you can use a negative step value. For example, if you want to get every third element from the end of the list, you can use a step of -3.
print(list_item[::-3])
```

# List Methods

```python
# append() method is used to add an element at the end of the list
list_item.append("New Item")
```

```python
list_item
```

```python
# insert() method is used to add an element at a specific index in the list. The syntax for insert() method is list.insert(index, element). The index parameter specifies the position where the element should be inserted, and the element parameter is the value that you want to insert into the list. For example, if you want to insert an element at index 2, you can use the following code:
list_item.insert(2, "Inserted Item")
```

```python
list_item
```

```python
lists_numbers=[1,2,3,23,43,1,6,3,5,6,7,9]
```

```python
# remove() method is used to remove an element from the list. The syntax for remove() method is list.remove(element). The element parameter is the value that you want to remove from the list . it will remove first occurrence of the element from the list. For example, if you want to remove the element 3 from the list, you can use the following code:
lists_numbers.remove(3)
```

```python
lists_numbers
```

```python
# Pop() method is used to remove an element from the list based on the index. The syntax for pop() method is list.pop(index). The index parameter specifies the position of the element that you want to remove from the list. If you do not provide an index, it will remove the last element from the list. For example, if you want to remove the element at index 2, you can use the following code:
lists_numbers.pop(2)
```

```python
lists_numbers
```

```python
# index() method is used to find the index of the first occurrence of an element in the list. The syntax for index() method is list.index(element). The element parameter is the value that you want to find the index of in the list. For example, if you want to find the index of the element 6 in the list, you can use the following code:
index_value=lists_numbers.index(6)
```

```python
index_value
```

```python
# count() method is used to count the number of occurrences of an element in the list. The syntax for count() method is list.count(element). The element parameter is the value that you want to count in the list. For example, if you want to count the number of occurrences of the element 1 in the list, you can use the following code:
count_value=lists_numbers.count(1)
```

```python
count_value
```

```python
#sort() method is used to sort the elements of the list in ascending order. The syntax for sort() method is list.sort(). For example, if you want to sort the list of numbers, you can use the following code:
lists_numbers.sort()
```

```python
lists_numbers
```

```python
# clear() method is used to remove all the elements from the list. The syntax for clear() method is list.clear(). For example, if you want to clear all the elements from the list, you can use the following code:
lists_numbers.clear()
```

```python
lists_numbers
```

```python
# reverse() method is used to reverse the order of the elements in the list. The syntax for reverse() method is list.reverse(). For example, if you want to reverse the order of the elements in the list, you can use the following code:
lists_numbers.reverse()
```

```python
lists_numbers
```

```python
# enumerate() method is used to get the index and value of each element in the list. The syntax for enumerate() method is enumerate(list). For example, if you want to get the index and value of each element in the list, you can use the following code:
for index, value in enumerate(lists_numbers):
    print(f"Index: {index}, Value: {value}")
```

## List Comprehension 

### we are going to use the looping statement inside the list so we can store the valuess inside the list in an iterable format and in list comprehension conditional statement are allowed as well

### Sytax for the comprehension is:
 ### [expression for item in iterable if condition]

### expression ---> what we want to print 
### for item in iterable ---> looping statements
### if condition ---> condional statement

```python
# i need to store values from 1 to 50 inside the list

list_items=[x for x in range(1,51)]
```

```python
list_items
```

```python
list_squared=[items**2 for items in range(1,16)]
```

```python
list_squared
```

```python
# if we only interested in looking over the squae values which are even

list_items=[items**2 for items in range (1,19) if items%2==0]
```

```python
list_items
```
