# Nested Dictionary 

## Nested dictionary willl be used for keeping a dictionary inside another dictionary


```python
#I want to store the inforamtion of multiple students like their name, marks , grade
Nested_dictionary={"Student_1":{"Name":"Santosh","Marks":85,"Grade":"A"},
                     "Student_2":{"Name":"Chetan","Marks":75,"Grade":"B"},
                     "Student_3":{"Name":"Neel","Marks":65,"Grade":"C"}}
```

```python
Nested_dictionary
```

```python
# Iterate through the nested dictionary
for outer_key,outer_value in Nested_dictionary.items():
    print(f"Outer key:{outer_key}and Outer value:{outer_value}")
    for key_1,value_1 in outer_value.items():
        print(f"Inner key:{key_1} and Inner Value:{value_1}")
```

## Dictioanry Comprehension 
### Dictioanry comprehension helps us with using looping statementalong with condtional statement inside our dictionary

## Syntax for dictioanry comprehension 
### {expression for item in iterable if conditon }

```python
comprehensive={x**2 for x in range(11)}
```

```python
comprehensive
```

```python
{x:x**2 for x in range(11)}
```

```python
even_dictionary={x:x**2 for x in range(11) if x%2==0}
```
